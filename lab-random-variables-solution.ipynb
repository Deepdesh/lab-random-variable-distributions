3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
def log_transform_clean(row):
    if np.isfinite(row) and row!=0:
        return np.log(row)
    else:
        return row
```

```python
numerical_dfc = numerical_df


def apply_log_trans(df):
    col_names_c = df.columns.tolist()
    for i in range(len(col_names_c)):
        #display(pd.Series(df.iloc[:,i].apply(log_transform_clean)).hist())
        #plt.show()
        df = pd.Series(df.iloc[:,i].apply(log_transform_clean))
        return df
def plot_log_trans(df):
    col_names_c = df.columns.tolist()
    for i in range(len(col_names_c)):
        print('for variable',col_names_c[i])
        display(pd.Series(df.iloc[:,i].apply(log_transform_clean)).hist())
        plt.show()
        
```

```python
apply_log_trans(numerical_dfc)
plot_log_trans(numerical_dfc)
```

Observation: Applying log transformation we see it was not beneficial for all of them. Caveat: Customer Lifetime Value and Total Claim Amount variables are now transformed and follow normal distribution. But it did not work for all of them.

No significant different observed. Perhaps, removing outliers would help. Trying that out in next step.

```python
def plot_box(df):
    col_names_c = df.columns.tolist()
    for i in range(len(col_names_c)):
        print('for variable',col_names_c[i])
        sns.boxplot(df.iloc[:,i])
        plt.show()
        
plot_box(numerical_dfc1)
```

```python
        
def remove_outlier(df):
    col_names_co = df.columns.tolist()
    for i in range(len(col_names_co)):

        iqr = np.percentile(df.iloc[:,i],75) - np.percentile(df.iloc[:,i],25)
        #iqr = np.percentile(df.['income'],75) - np.percentile(df.['income'],25)
        upper_limit = np.percentile(df.iloc[:,i],75) + 1.5*iqr
        
        extraordinary_points = df[df.iloc[:,i]>upper_limit]
        #len(extraordinary_points)
        df=df[df.iloc[:,i]<upper_limit]
        return df
```

```python
numerical_dfc2 = numerical_df
#apply_log_trans(numerical_dfc2)
remove_outlier(numerical_dfc2)
plot_log_trans(numerical_dfc2)
```

```python
numerical_dfc3 = numerical_df
#apply_log_trans(numerical_dfc2)
remove_outlier(numerical_dfc3)
plot_log_trans(numerical_dfc3)
```

From these observations, it is clear that applying same kind of transformations or cleaning processes don't work for each of these variables in the intended way. Perhaps, it is better to apply different ones for different variables. For instance, removing outliers seems to have turned 'Total Claim Amount' into a normal distribution but it didn't work for others.
